The research group at UBC need to target their recruitment efforts and manage the number of players they attract with the limited resources.

players.csv table has 196 observations with 7 variables. experience, hashedEmail, and name belong to character. subscribe belongs to logical, and played_hours and Age belong to double.

Experience indicates that how a player is experienced with the video games.
hashedEmail identifies the player's name in the sessions table.
subscribe indicates whether a player is subscribing the video game and played_hours indicates how much hours the player was in the game.

sessions.csv has 1535 observations with 5 variables. hashedEmail, start_time, and end_time belong to character and original_start_time and original_end_time belong to double.

start_time means the time when the player connected the server or logging started. end_time means the time when the player disconnected.
original_start_time is the raw timestamps from the source logs, written in epoch milliseconds. 1.71977e+12 is approximately equivalent to 30rd Jun, 2024. 

For players, name column is not important for prediction.

For sessions, large decimals numbers and epoch milliseconds make it difficult to interprete. The table is not tidy given that there are multiple measurements in the start_time and end_time columns. The time data was showing repeatedly by two overlapped columns.

kknn is an appropriate model for finding out target customers (k nearest points)
kknn detects non-linear patterns probabilities of being a large contributor with the various labels like gender, experience, and age.

For kknn, training set and testing set should have same proportion for the labels. For example, if 80% of the testing set consists of male, 80% of the training set also shold consist of male. Testing set cannot be used when building classfier. Standardizing is required.

Missing  data that can be observed in the dataset below. Removing them prior to building a model with drop_na() function solves this problem.
Class imbalance means one label is much more common than another. step_unsample function solves this problem.

If I want to predict classes of true for subscribe,
I load the data and covert the variable with as_factor and fct_recode, selecting subscribe(target), played_hours, gender, and age columns. 
I preprocess the data with recipe, step_normalize function.
I use nearest_neighbor, set_engine, set_mode, workflow() to create a model and predict the class of a new observation with predict function.

I split the data into 80/20 proportion, which means that 80% of the original data would end up in the training set with strata = subscribe. Then, I use training and testing function. This will be done with creatig a train prior to preprocess.

There will be cross validation with vfold_cv function to find the optimal k.

```r
new_obs <- tibble(Age = 44, played_hours = 30)

players_subscribe <- players |>
mutate(subscribe = factor(subscribe, levels = c(FALSE, TRUE))) |>
select(subscribe, Age, played_hours)

players_rec <- recipe(subscribe ~ Age + played_hours, data = players_subscribe) |>
step_normalize(all_numeric_predictors())

players_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 5) |>
set_engine("kknn") |>
set_mode("classification")

players_wf <- workflow() |>
add_recipe(players_rec) |>
add_model(players_spec) |>
fit(data = players_subscribe)

players_predict <- predict(players_wf, new_data = new_obs, type = "class")
players_predict
```
